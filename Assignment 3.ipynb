**Q.1 DATA CLEANING & PREPROCEESING**

   Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

  * Load the dataset

```python
df = pd.read_csv(r"C:\Users\RD\OneDrive\Desktop\Pandas\Matplotlib\Assignments\train.csv")
df
```

* Check  for missing values

```python
df.isnull().sum()
```

* Fill missing Age values with median or mean

```python
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df.isnull().sum()
```

Now there is no missing values in Age column

* Handle missing Embarked values by filling with the most common port

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

So now we can see that there is no missing values in 'Embarked' column

* Drop unnecessary columns like Name , Ticket ,Cabin for visualization purposes

**Before Dropping Columns**

```python
df
```

```python
df.drop(['Name','Ticket','Cabin'],axis=1,inplace=True)
```

```python
df
```

Now we drop a column  **Name , Ticket ,Cabin**

**2. Univariate Analysis**

* PLot the **distribution of Age** using a histogram and KDE plot

```python
sns.histplot(df['Age'], kde=True)
plt.title("Distribution of Age")
plt.show()
```

From this histogram we can simply analyze that most of the passenger between the age group 20-40 and the highest bar between the 28-30 age group

* Plot the **count of passengers** by Sex using a bar chart

```python
sns.countplot(x='Sex',data=df)
plt.title("Count of Passengers by Sex")
plt.show()
```

From this barchart we can simply analyze that :-

Male passengers are more compared to Female passenger.

* Show the **distribution of Fare** using a boxplot

```python
sns.boxplot(x='Fare',data=df)
plt.title("Fare distribution")
plt.plot()
```

From this we can simply analyze that most of the passengers paid lower fare

**3. Bivariate Analysis**

* Compare survival rate across **Sex** using a countplot.

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival count by sex")
plt.plot()
```

from this  countplot we analyze that :-
The female had a higher survival rate than Male 

* Compare **Pclass vs Fare** using a boxplot.

```python
sns.boxplot(x='Pclass' ,y='Fare', data=df)
plt.title("Pclass vs Fare")
plt.plot()
```

From this we analyze the :-

First class passenger paid the higher fare

Second class passenger paid the moderate fare

and third class passenger paid lowest fare among all.

* Visualize **Age vs Fare** using a scatter plot,color-coded by Survived. 

```python
sns.scatterplot(x='Age', y='Fare',hue='Survived', data=df)
plt.title("Age Vs Fare (color coded by survived)")
plt.plot()
```

The scatter plot shows that the passenger who paid higher fares had a higher Survival rate and passenger which paid fare below 100 are mostly unsurvived between 20 - 40 age group .
